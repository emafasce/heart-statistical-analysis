```python
#Script used for finding theoretical distribution similar to empirical ones

import warnings
import numpy as np
import pandas as pd
import scipy.stats as st
import statsmodels.api as sm
import matplotlib
import matplotlib.pyplot as plt


# Create models from data
def best_fit_distribution(data, bins=200, ax=None):
    """Model data by finding best fit distribution to data"""
    # Get histogram of original data
    y, x = np.histogram(data, bins=bins, density=True)
    x = (x + np.roll(x, -1))[:-1] / 2.0

    # Distributions to check
    DISTRIBUTIONS = [        
         st.norm,st.uniform, st.gamma,
    ]

    # Best holders
    best_distribution = st.uniform
    best_params = (0.0, 1.0)
    best_sse = np.inf

    # Estimate distribution parameters from data
    for distribution in DISTRIBUTIONS:

        # Try to fit the distribution
        try:
            # Ignore warnings from data that can't be fit
            with warnings.catch_warnings():
                warnings.filterwarnings('ignore')

                # fit dist to data
                params = distribution.fit(data)

                # Separate parts of parameters
                arg = params[:-2]
                loc = params[-2]
                scale = params[-1]

                # Calculate fitted PDF and error with fit in distribution
                pdf = distribution.pdf(x, loc=loc, scale=scale, *arg)
                sse = np.sum(np.power(y - pdf, 2.0))

                # if axis pass in add to plot
                try:
                    if ax:
                        pd.Series(pdf, x).plot(ax=ax)
                    end
                except Exception:
                    pass

                # identify if this distribution is better
                if best_sse > sse > 0:
                    best_distribution = distribution
                    best_params = params
                    best_sse = sse

        except Exception:
            pass

    return (best_distribution.name, best_params)

def make_pdf(dist, params, size=10000):
    """Generate distributions's Probability Distribution Function """

    # Separate parts of parameters
    arg = params[:-2]
    loc = params[-2]
    scale = params[-1]

    # Get sane start and end points of distribution
    start = dist.ppf(0.00001, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.00001, loc=loc, scale=scale)
    end = dist.ppf(0.99999, *arg, loc=loc, scale=scale) if arg else dist.ppf(0.99999, loc=loc, scale=scale)

    # Build PDF and turn into pandas Series
    x = np.linspace(215, 275, size)
    y = dist.pdf(x, loc=loc, scale=scale, *arg)
    pdf = pd.Series(y, x)

    return pdf
```

```python
#LEAVE ONE OUT 
from sklearn.metrics import recall_score
from sklearn.model_selection import LeaveOneOut
import numpy as np
%matplotlib inline
from sklearn import tree
import matplotlib.pyplot as plt

from sklearn.model_selection import ParameterGrid
from sklearn.metrics import accuracy_score


loocv=LeaveOneOut()
param_grid = {'criterion': ['gini', 'entropy'], 'max_features': ['auto', 'sqrt', 'log2'], 'splitter':['best', 'random'], 
              'max_depth':[3,4,5,6,7, None]}

acc_loo=[]
i=0
for comb in list(ParameterGrid(param_grid)):
    acc_total=0
    for train_index, test_index in loocv.split(df_dummy, illness):
        X_train, X_test = df_dummy.loc[train_index], df_dummy.loc[test_index]
        y_train, y_test=illness[train_index], illness[test_index]
        decision=tree.DecisionTreeClassifier(criterion=comb['criterion'], max_depth=comb['max_depth'], max_features=comb['max_features'], splitter=comb['splitter'])
        decision.fit(X_train, y_train)
        pred=decision.predict(X_test)
        acc=accuracy_score(pred, y_test)
        acc_total+=acc
    acc_loo.append(acc_total/len(df_dummy))
    i=i+1
    print(i)

#STRATIFIED K FOLD

from sklearn.model_selection import StratifiedKFold

acc_kfold=[]
skfold=StratifiedKFold(n_splits=20)

i=0
for comb in list(ParameterGrid(param_grid)):
    acc_total=0
    for train_index, test_index in skfold.split(df_dummy, illness):
        X_train, X_test = df_dummy.loc[train_index], df_dummy.loc[test_index]
        y_train, y_test=illness[train_index], illness[test_index]
        decision=tree.DecisionTreeClassifier(criterion=comb['criterion'], max_depth=comb['max_depth'], max_features=comb['max_features'], splitter=comb['splitter'])
        decision.fit(X_train, y_train)
        pred=decision.predict(X_test)
        acc_total+=np.sum(pred==y_test)
    acc_kfold.append(acc_total/len(df_dummy))
    print(i)
    i=i+1


winner_loo=list(ParameterGrid(param_grid))[np.argmax(acc_loo)]
winner_kfold=list(ParameterGrid(param_grid))[np.argmax(acc_kfold)]

decision=tree.DecisionTreeClassifier(criterion=winner_loo['criterion'], max_depth=winner_loo['max_depth'], max_features=winner_loo['max_features'], splitter=winner_loo['splitter'])
decision.fit(train, illness_train)
pred=decision.predict(test)
decision_loo_acc=accuracy_score(pred,illness_test)
decision_loo_rec=recall_score(pred, illness_test)
print("acc:",accuracy_score(pred,illness_test))
print("recall:",recall_score(pred, illness_test))
#acc_loo=accuracy_score(pred,illness_test)

decision=tree.DecisionTreeClassifier(criterion=winner_kfold['criterion'], max_depth=winner_kfold['max_depth'], max_features=winner_kfold['max_features'], splitter=winner_kfold['splitter'])
decision.fit(train, illness_train)
pred=decision.predict(test)
decision_kfold_acc=accuracy_score(pred,illness_test)
decision_kfold_rec=recall_score(pred, illness_test)
print("acc:",accuracy_score(pred,illness_test))
print("recall:",recall_score(pred, illness_test))
#acc_kfold=accuracy_score(pred,illness_test)
```

```python
import plotly
from plotly.offline import plot
import plotly.graph_objs as go 

def plot3d(X_reduced, illness):
    trace1 = go.Scatter3d(
        x=X_reduced[:,0],
        y = X_reduced[:,1],
        z = X_reduced[:,2],
        mode='markers',
        marker=dict(
            color=illness,
            colorscale=[(0,"#0085ff"), (1,"#00ffab")],
            size=5,               
            opacity=1))

    data = [trace1]
    layout = go.Layout(
        xaxis=dict(
            title='PCA',
            titlefont=dict(
               family='Courier New, monospace',
               size=18,
               color='#7f7f7f'
           )
       )
    )
    fig = go.Figure(data=data, layout=layout)
    plot(fig, filename='3D PCA')
```
